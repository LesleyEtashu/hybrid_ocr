# pdfplumber

pdfplumber is a Python library (not an AI model) used to extract text, tables, and layout information directly from PDF files. It works by reading embedded text from the PDF structure, which means it is suitable for digitally generated PDFs, but not for scanned documents that require OCR.

```python
import pdfplumber

# Extract text and tables
results = []
with pdfplumber.open(pdf_path) as pdf:
    for i, page in enumerate(pdf.pages):
        page_content = []

        # Extract text
        text = page.extract_text() or ""
        if text:
            page_content.append(text)

        # Extract tables
        tables = page.extract_tables()
        for j, table in enumerate(tables):
            # Convert table to markdown format
            if table and len(table) > 0:
                md_table = []
                # Header
                header = table[0]
                md_table.append("| " + " | ".join(str(cell or "") for cell in header) + " |")
                md_table.append("| " + " | ".join("---" for _ in header) + " |")
                # Rows
                for row in table[1:]:
                    md_table.append("| " + " | ".join(str(cell or "") for cell in row) + " |")
                page_content.append("\n".join(md_table))

        results.append("\n\n".join(page_content))
        print(f"Page {i+1}: {len(text)} chars, {len(tables)} tables")

full_text = "\n\n---\n\n".join(results)
print(f"\n Total: {len(full_text):,} chars")

```

```python
# Save results
output_path = "pdfplumber_output.md"
with open(output_path, "w", encoding="utf-8") as f:
    f.write(full_text)

print(f"Saved to: {output_path}")
files.download(output_path)
```
