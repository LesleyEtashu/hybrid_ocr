```python
#PDF → images → PaddleOCR text
from pdf2image import convert_from_path
from paddleocr import PaddleOCR
import numpy as np

print("Using PDF:", pdf_path)

# Init OCR (same as step 4)
ocr = PaddleOCR(lang="en")

#  Convert PDF to images
pages = convert_from_path(pdf_path, dpi=150)
print("Pages loaded:", len(pages))

all_pages_text = []

# OCR each page
for i, page in enumerate(pages):
    print(f"OCR on page {i+1}/{len(pages)}...")
    img = np.array(page)

    # Run OCR on the page image
    result = ocr.ocr(img)

    page_lines = []

    # result is usually a list (one element per page/image)
    for entry in result:
        # dict-style modern output (PaddleOCR 3.x)
        if isinstance(entry, dict) and "data" in entry:
            for det in entry["data"]:
                text = det.get("text", "")
                if text:
                    page_lines.append(text)

        #  classic list/tuple style output
        elif isinstance(entry, (list, tuple)):
            for item in entry:
                if isinstance(item, (list, tuple)) and len(item) >= 2:
                    text_info = item[1]
                    if isinstance(text_info, (list, tuple)) and len(text_info) >= 1:
                        text = text_info[0]
                    else:
                        text = str(text_info)
                    if text:
                        page_lines.append(text)

    # Join lines for this page
    all_pages_text.append("\n".join(page_lines))

#  Join all pages into one big string
paddle_text = "\n\n".join(all_pages_text)

# Save to file
with open("paddle.txt", "w", encoding="utf-8") as f:
    f.write(paddle_text)

print("Saved PaddleOCR output to paddle.txt, length:", len(paddle_text))
```

```python
from pdf2image import convert_from_path
from paddleocr import PaddleOCR
import numpy as np

ocr = PaddleOCR(lang="en")
pages = convert_from_path(pdf_path, dpi=150)

with open("paddleOCR_output.md", "w", encoding="utf-8") as f:
    f.write("# PaddleOCR Output\n\n")
    f.write("Source PDF: aa602a2b-0cf0-5d8f-8050-4e73aa4e3d6c_2.pdf\n")
    f.write("Pages processed: 68\n\n---\n")

    for i, page in enumerate(pages, start=1):
        f.write(f"\n## Page {i}\n")
        img = np.array(page)
        result = ocr.ocr(img)

        page_text = []

        if result:
            for block in result:
                if isinstance(block, list):
                    for line in block:
                        if len(line) >= 2:
                            page_text.append(line[1][0])

        if page_text:
            f.write("\n".join(page_text) + "\n")
        else:
            f.write("*(No text detected)*\n")

        f.write("\n---\n")
```
